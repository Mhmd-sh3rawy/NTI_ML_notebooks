```python
import pandas as pd
df_bank = pd.read_csv('BankChurners.csv')
df_bank.head()
```

```python
display(df_bank.columns)
display(df_bank.info())
display(df_bank.describe())
df_bank.hist()
```

```python
precomputed_cols = df_bank.columns[-2:]
drop_cols = ['CLIENTNUM', 'Attrition_Flag', precomputed_cols[0], precomputed_cols[1]]
```

```python
X = df_bank.drop(drop_cols, axis=1)
y = df_bank['Attrition_Flag']

X_cat = X.select_dtypes('object')
X_num = X.select_dtypes(['float64', 'int64'])
```

```python
X_num.columns
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

cat_pipeline = Pipeline([
    ('one_hot_encoder', OneHotEncoder())
])

X_cat_prepared = cat_pipeline.fit_transform(X_cat)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

num_pipeline = Pipeline([
    ('std_scaler', StandardScaler())
])

X_num_prepared = num_pipeline.fit_transform(X_num)
```

```python
X_prepared = pd.concat([pd.DataFrame(X_num_prepared), pd.DataFrame(X_cat_prepared.toarray())], axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_prepared, y, test_size=0.2, random_state=42)
```

```python
#from imblearn.over_sampling import SMOTE

#smote = SMOTE()
#X_train, y_train = smote.fit_resample(X_train, y_train)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB

mdl = BernoulliNB()
mdl.fit(X_train, y_train)
mdl.score(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

y_pred = mdl.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
