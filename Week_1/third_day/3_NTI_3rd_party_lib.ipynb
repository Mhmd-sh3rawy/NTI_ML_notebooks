```python
salary['IsManager'] = salary['Job_Title'].str.contains('Manager')
salary['IsManager']
```

```python
salary['ExpLevel'] = pd.cut(salary['Years_of_Experience'],
                            bins=[0, 2, 5, 10, 20, np.inf],
                            labels=['Entry', 'Junior', 'Mid', 'Senior', 'Veteran'])
salary['ExpLevel']
```

```python
salary['Salary_stdz'] = (salary['Salary'] - salary['Salary'].mean())/salary['Salary'].std()
salary['Salary_stdz']
```

```python
salary['GenderBin'] = salary['Gender'].map({'Male': 0, 'Female': 1})
salary['GenderBin']
```

```python
engineer = salary['Job_Title'].str.contains('Engineer')
analyst = salary['Job_Title'].str.contains('Analyst')
scientist = salary['Job_Title'].str.contains('Scientist')


salary['JobType'] = np.select(condlist=[engineer, analyst, scientist],
                              choicelist = ['Engineer', 'Analyst', 'Scientist'],
                              default='Other')

salary[['Job_Title','JobType']]
```
